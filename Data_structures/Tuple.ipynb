## Tuples in Python

Tuples are ordered collections of heterogeneous data that are unchangeable. Heterogeneous means tuple can store variables of all types.

Tuple has the following characteristics:





```python
# Creating tuple using tuple constructor

n_tuple = (10,20,30.5)
print(n_tuple)
```

## Packing and Unpacking
A tuple can also be created without using a tuple() constructor or enclosing the items inside the parentheses. It is called the variable Packing.

```python
#Variable packing
tuple1 = 1,2,3,"hello"
print(tuple1)

# Variable Unpacking
i,j,k,l = tuple1
print(i,j,k,l)
```

## Adding and changing items in a Tuple

A list is a mutable type, which means we can add or modify values in it, but tuples are immutable, so they cannot be changed.

Also, because a tuple is immutable there are no built-in methods to add items to the tuple.

If you try to modify the value you will get an error.

**Note:**

To do so type cast tuple to list and again to tuple after changes.

```python
# Copying a tuple
# We can create  acopy of tuple using assignment operator "=".
# This will create e reference copy and a deep copy is not possible because tuples are immutable.


tuple2 = tuple1
print(tuple2)
```

```python
# Sum function
print(sum(tuple1))

# Max and min functions
print("max:", max(tuple1))
print("min:", min(tuple1))
```

## When to use tuple?

As tuples and lists are similar data structures, and they both allow sequential data storage, tuples are often referred to as immutable lists. So the tuples are used for the following requirements instead of lists.

1. There are no append() or extend() to add items and similarly no remove() or pop() methods to remove items. This ensures that the data is write-protected. As the tuples are Unchangeable, they can be used to represent read-only or fixed data that does not change.

2. As they are immutable, they can be used as a key for the dictionaries, while lists cannot be used for this purpose.

3. As they are immutable, the search operation is much faster than the lists. This is because the id of the items remains constant.

4. Tuples contain heterogeneous (all types) data that offers huge flexibility in data that contains combinations of data types like alphanumeric characters.
