## Python Lists

Python list is an ordered sequence of items.





The following are the properties of a list.

**Mutable**: The elements of the list can be modified. We can add or remove items to the list after it has been created.

**Ordered**: The items in the lists are ordered. Each item has a unique index value. The new items will be added to the end of the list.

**Heterogenous**: The list can contain different kinds of elements i.e; they can contain elements of string, integer, boolean, or any type.

**Duplicates**: The list can contain duplicates i.e., lists can have two items with the same values.


**Why use a list?**

-- The list data structure is very flexible It has many unique inbuilt functionalities like pop(), append(), etc 
which makes it easier, where the data keeps changing.

-- Also, the list can contain duplicate elements i.e two or more items can have the same values.

-- Lists are Heterogeneous i.e, different kinds of objects/elements can be added

-- As Lists are mutable it is used in applications where the values of the items change frequently.


```python
# Creating The list

# list constructor
ls1 = list((1,2,3,4,5,6))

# using square brackets[]
ls2 = [1,2,3,4]

# empty list using list constructor
ls3 = list()
 
# empty list using square brackets[]
ls4 = []
```

```python
# Accessing elements of list

# element by index
print(ls1[2])

# slicing the list
print(ls1[2:5])
```

```python
# Modifying the list

# Appending an element
my_list = [100,200,300,454,689]
my_list.append(8000)
print(my_list)

# extending list into list
my_list.extend(ls1)
print(my_list)

#Inserting an element at specfic index
my_list.insert(1,5)
print(my_list)

# removing an element by value
my_list.remove(5)

# removing an elemnt by index
del my_list[1]
print(my_list)

# popping an element
my_list.pop(1)
print(my_list)

# clearing a list
ls2.clear()

# Updating element by index
my_list[1] = 4000

# reversing a list 
my_list.reverse()

# sorting a list
my_list.sort()
print(my_list)
```

```python
# Check if element is in list

if 4000 in my_list:
    print("Found")
```

```python
# list operations

# length of list
print(len(my_list))

# copying the list


# Copy() function creates a shallow copy means changes in original list do not reflect in copied list
copied_my_list = my_list.copy()


# Deep copying is where  changes happen in original and copied version.

nl = my_list 
print(nl)

# concatenating list 
 
concated_list  = nl + my_list
print(concated_list) 


# counting occurence of elemnts in list

print(concated_list.count(2))

# finding index of elements
print(nl.index(100))
```
